```python
import numpy as np

# データセットを準備します。ここでは、n*100次元のデータを生成しています。
n = 3072  # ここでは例としてnを5としています。実際には、必要な値に設定してください。
data = np.random.rand(400, n)  # 1000個のn*100次元データを生成

print(f'{data.shape=}')

# 分散共分散行列を計算します。
cov_matrix = np.cov(data, rowvar=False)

print(cov_matrix.shape)
```
